```python
codes = {
    'LVA': 'LV',
    'MLT': 'MT',
    'CYP': 'CY',
    'DNK': 'DK',
    'GRC': 'GR',
    'ITA': 'IT',
    'BEL': 'BE',
    'CZE': 'CZ',
    'HRV': 'HR',
    'SWE': 'SE',
    'EST': 'EE',
    'DEU': 'DE',
    'FIN': 'FI',
    'LTU': 'LT',
    'ESP': 'ES',
    'LUX': 'LU',
    'BGR': 'BG',
    'POL': 'PL',
    'ROU': 'RO',
    'AUT': 'AT',
    'SVK': 'SK',
    'NLD': 'NL',
    'IRL': 'IE',
    'FRA': 'FR',
    'HUN': 'HU',
    'PRT': 'PT',
    'SVN': 'SI'
}

economic_level['CountryCode'] = economic_level['Code'].map(codes)
```

```python
economic_level = economic_level[economic_level['Code'].isin(['LVA', 'MLT', 'CYP', 'DNK', 'GRC', 'ITA', 'BEL', 'CZE', 'HRV', 'SWE',
 'EST', 'DEU', 'FIN', 'LTU', 'ESP', 'LUX', 'BGR', 'POL', 'ROU', 'AUT',
 'SVK', 'NLD', 'IRL', 'FRA', 'HUN', 'PRT', 'SVN'])] 
economic_level = economic_level[economic_level['Year']>= 2013]
```

```python
economic_level
```

```python
merged_df = lung_cancer.merge(air_pollution, left_on=['CountryCode', 'year'], right_on=['CountryCode', 'year'], how='inner')
merged_df = merged_df[merged_df['year']>2019] 
```

```python

final_df = merged_df[['id', 'age', 'gender', 'country', 'year', 'cancer_stage', 'hypertension', 'asthma', 'cirrhosis', 'treatment_type', 'survived', 'Max_AQI', 'Air_Quality', 'CountryCode']]
```
